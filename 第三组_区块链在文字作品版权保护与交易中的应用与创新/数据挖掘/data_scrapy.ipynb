```python
import time
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
import pandas as pd
from selenium.webdriver.common.keys import Keys
import csv
from selenium.webdriver.common.action_chains import ActionChains
import numpy as np

options = Options()

print(zuopin)


    
driver = webdriver.Firefox(executable_path='D:\Install\Anaconda3\geckodriver.exe', firefox_options=options)
driver.set_window_size('1500', '800')
print('browser ready')

content=[]

url = 'http://product.dangdang.com/25235726.html'
driver.get(url)
time.sleep(3)#程序暂停运行，让网页加载一会儿
i=1
for l in zuopin:
    
    driver.find_element_by_xpath('//*[@id="key_S"]').send_keys(Keys.CONTROL,'a')
    time.sleep(2)
    driver.find_element_by_xpath('//*[@id="key_S"]').send_keys(l)
    time.sleep(2)#程序暂停运行，让网页加载一会儿
    driver.find_element_by_xpath('//*[@id="key_S"]').send_keys(Keys.ENTER)
    time.sleep(5)
    
    driver.find_element_by_xpath('//*[@class="pic"]').click()
    time.sleep(4) 
    driver.switch_to_window(driver.window_handles[i])
        
    ele=driver.find_element_by_xpath('//*[@id="content"]')
    data=ele.text
    data=data.split('\n')
    d=''
    for line in data:
        if line != '内容简介':
            if line != '':
                if line != '显示全部信息':
                    d=d+line
    
    content.append(d)
    
    i=i+1
    time.sleep(4)
    
print(content)
    
```
